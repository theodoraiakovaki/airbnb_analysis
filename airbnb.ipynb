```python
import altair as alt 
from ipywidgets import VBox, HTML, Output, Dropdown, Checkbox, HBox, interactive
from IPython.display import display, Markdown
from datetime import datetime, timedelta

cities = {
    'Athens': athens,
    'Amsterdam': amsterdam,
    'Barcelona': barcelona,
    'Berlin': berlin,
    'Copenhagen': copenhagen,
    'Dublin': dublin,
    'Lisbon': lisbon,
    'London': london,
    'Madrid': madrid,
    'Paris': paris,
    'Rome': rome,
    'Venice': venice,
    'Vienna': vienna,
}

output_plot = Output()
output_description = Output()
output_percentages = Output()
output_description_activity = Output()
output_activity_averages = Output()
output_activity = Output()
output_total_listings = Output()
output_license = Output()
output_description_license = Output()
output_percentages_license = Output()
output_short_term = Output()
output_description_short_term = Output()
output_short_term_percentages = Output()
output_listings_per_host = Output()
output_description_listings = Output()
output_listings_per_host_percentages = Output()
output_hosts_table = Output()
        

def plot_room_type(selected_city, only_entire_homes):
    with output_plot:
        output_plot.clear_output()
        room_type_counts = cities[selected_city]['room_type'].value_counts().reset_index()   # Get room type counts for the selected city
        room_type_counts.columns = ['Room_type', 'Listings']
        # If only_entire_homes_checkbox is selected, filter for 'Entire home/apt' listings
        if only_entire_homes:
            all_room_types = ['Entire home/apt', 'Private room', 'Shared room', 'Hotel room']
            filtered_counts = room_type_counts[room_type_counts['Room_type'] == 'Entire home/apt']
            complete_counts = pd.DataFrame({'Room_type': all_room_types})
            room_type_counts = pd.merge(complete_counts, filtered_counts[['Room_type', 'Listings']], on='Room_type', how='left').fillna(0)
            room_type_counts['Listings'] = room_type_counts['Listings'].astype(int)
        else:
            room_type_counts['Listings'] = room_type_counts['Listings'].astype(int)
        total_listings_filtered = room_type_counts['Listings'].sum()
        room_type_counts['Percentage'] = (room_type_counts['Listings'] / total_listings_filtered  * 100).round(1)
        color_scale = alt.Scale(
            domain=['Entire home/apt', 'Private room', 'Shared room', 'Hotel room'],
            range=['#F44336', '#4CAF50', '#03A9F4', '#9E9E9E'] 
        )
        chart = alt.Chart(room_type_counts).mark_bar().encode(
          y=alt.Y('Room_type:O', sort='-x', title=None),
          x=alt.X('Listings:Q', title='listings',
                     axis=alt.Axis(tickCount=3, values=[0, 5000, 10000])),
            tooltip=[
                alt.Tooltip('Room_type:N', title='Room Type'),
                alt.Tooltip('Listings:Q', title='Number of Listings'),
                alt.Tooltip('Percentage:Q', title='Percentage')
            ],
         color=alt.Color('Room_type:N', scale=color_scale, legend=None)
        ).properties(
            width=400,  
            height=300 
        ).configure_view(
            stroke=None  
        )

        display(chart)


    with output_description:
        output_description.clear_output()
        description_text = (
            "Airbnb hosts can list<br> entire homes/apartments, private,<br> shared rooms, and more<br> recently hotel rooms.<br>"
            "Depending on the room<br> type and activity, a<br> residential Airbnb listing<br> could be more like a hotel,<br>"
            "disruptive for neighbours,<br> taking away housing,<br> and illegal."
        )
        display(Markdown(description_text))
    

    with output_percentages:
       output_percentages.clear_output() 
       largest = room_type_counts.iloc[0]
       largest_listings_formatted = f"{largest['Listings']:,}" 
       percentage_text = (
        f"<h2 style='margin: 0;'><strong>{largest['Percentage']}%</strong></h2>"
        f"<p style='font-size: 18px; margin: 0;'>{largest['Room_type'].replace('/', '/')}</p>"
       )
    
       for index, row in room_type_counts.iterrows():
            listings_formatted = f"{row['Listings']:,}" 
            percentage_text += (
                f"<p style='margin: 5px 0;'><strong>{listings_formatted} ({row['Percentage']}%)</strong><br>"
                f"{row['Room_type'].replace('/', '/')}</p>"
            )

       display(Markdown(percentage_text))
        
def plot_activity(selected_city, activity):
    with output_activity:
        output_activity.clear_output()
        current_date = datetime.now()
        selected_city_data = cities[selected_city]
        selected_city_data['last_review'] = pd.to_datetime(selected_city_data['last_review']) # Convert 'last_review' column to datetime to filter recent listings
        filtered_data = selected_city_data.copy()
        # If 'activity_checkbox' is checked, filter for active listings with recent reviews and high bookings
        if activity_checkbox.value:
            six_months_ago = current_date - timedelta(days=180)
            recent_reviews = filtered_data[filtered_data['last_review'] >= six_months_ago]
            frequently_booked = recent_reviews[recent_reviews['total_bookings'] > 60]
            filtered_data = frequently_booked
        else: 
            filtered_data = selected_city_data

        booking_bins = [0, 1, 30, 60, 90, 120, 150, 180, 210, 240, 255, float('inf')]
        booking_labels = [
            "0", "1-30", "31-60", "61-90", "91-120", "121-150", 
            "151-180", "181-210", "211-240", "241-255", "255+"
        ]
        filtered_data = filtered_data.copy() 
        filtered_data['booking_range'] = pd.cut(filtered_data['total_bookings'], bins=booking_bins, labels=booking_labels, right=False)
        booking_counts = filtered_data['booking_range'].value_counts().sort_index().reset_index()
        booking_counts.columns = ['Nights_booked', 'Listings']
        activity_chart = alt.Chart(booking_counts).mark_bar(color='#1f77b4').encode(
            x=alt.X('Nights_booked:O', title='Occupancy (last 12 months)', sort=booking_labels, axis=alt.Axis(labelAngle=0)),
            y=alt.Y('Listings:Q', title='Listings'),
            tooltip=[
                alt.Tooltip('Nights_booked:N', title='Nights Booked'),
                alt.Tooltip('Listings:Q', title='Number of Listings')
            ]
        ).properties(
            width=400,
            height=200
        ).configure_axis(
            labelFontSize=8,
            titleFontSize=14
        ).configure_title(
            fontSize=16,
            fontWeight="bold"
        ).configure_view(
            stroke=None  
        )
        
        display(activity_chart) 
    with output_description_activity:
        output_description_activity.clear_output()
        description = (
                "The minimum stay, price and number of<br> reviews have been used to estimate the<br> number of"
                "<strong> nights booked</strong> and <strong>the income</strong><br> for each listing, for the last 12 months.<br>"
                "Is the home, apartment or room rented<br> frequently and displacing units of housing and<br>"
                "residents? Does the income from Airbnb</br> incentivise short-term rentals vs long-term</br> housing?"
            )
        display(Markdown(description))
    
    with output_activity_averages: # Display average statistics for nights booked, price, and income
            output_activity_averages.clear_output()
            average_nights_booked = filtered_data['total_bookings'].mean()  
            average_price = filtered_data['price'].mean() if 'price' in filtered_data.columns else 0  
            average_income = (filtered_data['price'] * filtered_data['total_bookings']).mean() if 'price' in filtered_data.columns else 0 
            average_info_html = (
                 f"<div style='text-align: right;'>"
                 f"<h2 style='margin: 0; font-size: 24px;'><strong>{average_nights_booked:.2f}</strong></h2>"
                 f"<p style='font-size: 14px; margin: 0;'>Average Nights Booked</p>"
                 f"<h2 style='margin: 10px 0 0; font-size: 24px;'><strong>€{average_price:,.2f}</strong></h2>"  # Added € symbol here
                 f"<p style='font-size: 14px; margin: 0;'>Average Price</p>"
                 f"<h2 style='margin: 10px 0 0; font-size: 24px;'><strong>€{average_income:,.2f}</strong></h2>"  # And here
                 f"<p style='font-size: 14px; margin: 0;'>Average Income</p>"
                 f"</div>"
            )
        
            display(HTML(average_info_html))

def plot_license(selected_city, cities, license):
    city_data = cities[selected_city].copy()
    city_data['license'] = city_data['license'].fillna(0)
    city_data['license'] = city_data['license'].apply(
        lambda x: 'licensed' if x != 0 and x not in ['Exempt'] 
        else ('exempt' if x == 'Exempt' else 'unlicensed')
    )
    
    license_counts = city_data['license'].value_counts()
    license_data = pd.DataFrame({
        'License_Status': ['unlicensed', 'licensed', 'exempt'],
        'Count': [
            license_counts.get('unlicensed', 0),
            license_counts.get('licensed', 0),
            license_counts.get('exempt', 0)
        ]
    })
    # If only unlicensed data is requested, zero out other counts
    if license:
        unlicensed_count = license_data[license_data['License_Status'] == 'unlicensed']['Count'].values[0]
        license_data['Count'] = [unlicensed_count, 0, 0] 
    # Calculate percentage for each category    
    total_count = license_data['Count'].sum() or 1  
    license_data['Percentage'] = (license_data['Count'] / total_count * 100).round(1)

    with output_license:
        output_license.clear_output()
        donut_chart = alt.Chart(license_data).mark_arc(innerRadius=90).encode(
            theta=alt.Theta(field='Count', type='quantitative', title='Count'),
            color=alt.Color(field='License_Status', type='nominal', title=None, 
                            scale=alt.Scale(domain=['licensed', 'exempt', 'unlicensed'], 
                                            range=['#1565C0', '#E53935', '#9E9E9E'])),
            tooltip=[
                alt.Tooltip('License_Status:N', title='License Status'),
                alt.Tooltip('Count:Q', title='Count'),
                alt.Tooltip('Percentage:Q', title='Percentage', format='.1f')
            ]
        ).properties(
            width=400,
            height=400
        )
        display(donut_chart)

    with output_description_license:
        output_description_license.clear_output()
        description = (
            "According to the law, specific short-term rentals<br> require a license.<br>"
            "How many listings are unlicensed, falsely<br> claiming exemptions, and how many do Airbnb<br> continue to advertise and profit from?"
        )
        display(HTML(description))

    with output_percentages_license:
        output_percentages_license.clear_output()
        license_info_html = "<div style='text-align: left;'>"
        for _, row in license_data.iterrows():
            license_info_html += (
                f"<h2 style='margin: 0; font-size: 24px;'><strong>{row['Percentage']}%</strong></h2>"
                f"<p style='font-size: 14px; margin: 0;'>{row['License_Status'].capitalize()}</p>"
                f"<p style='font-size: 16px; margin: 0;'>{row['Count']} ({row['Percentage']}%)</p>"
            )
        license_info_html += "</div>"

        display(HTML(license_info_html))

def plot_short_term(selected_city, cities, short_term):
       with output_short_term:
        output_short_term.clear_output()  
        city_data = cities[selected_city]
        if short_term:
            city_data = city_data[city_data['minimum_nights'] < 30]   
        min_nights_counts = city_data['minimum_nights'].value_counts().sort_index()
        min_nights_data = pd.DataFrame({
            'minimum_nights': min_nights_counts.index,
            'listings': min_nights_counts.values
        })

        min_nights_data['minimum_nights'] = min_nights_data['minimum_nights'].clip(upper=35)
        min_nights_data = min_nights_data.groupby('minimum_nights').sum().reset_index()
        short_term_chart = alt.Chart(min_nights_data).mark_bar(color='#4A90E2').encode(
            x=alt.X('minimum_nights:O', title='Minimum Nights', axis=alt.Axis(labelAngle=0)),
            y=alt.Y('listings:Q', title='Listings'),
            tooltip=[alt.Tooltip('minimum_nights:O', title='Minimum Nights'),
                     alt.Tooltip('listings:Q', title='Listings')]
        )

        display(short_term_chart)
           
       with output_description_short_term:
            output_description_short_term.clear_output()
            description = (
            'The housing policies of cities and towns can be restrictive of short-term rentals, to protect housing<br> for residents.'
            'By looking at the <strong> "minimum nights"</strong> setting for listings, we can see if the market has shifted to<br> longer-term stays. Was it to avoid regulations, or  in response to changes in travel demands?<br>'
            'In some cases, Airbnb has moved large numbers of their listings to longer-stays to avoid short-term<br> rental regulations and accountability.'
            )
            display(HTML(description))
           
       with output_short_term_percentages:
          output_short_term_percentages.clear_output()
          short_term_count = city_data[city_data['minimum_nights'] < 30].shape[0]
          longer_term_count = city_data[city_data['minimum_nights'] >= 30].shape[0]
          total_listings = short_term_count + longer_term_count
          short_term_percentage = (short_term_count / total_listings * 100) if total_listings > 0 else 0
          longer_term_percentage = (longer_term_count / total_listings * 100) if total_listings > 0 else 0
          short_term_percentage = round(short_term_percentage, 1)
          longer_term_percentage = round(longer_term_percentage, 1)
          output_html = (
            f"<div style='text-align: center; font-family: Arial, sans-serif;'>"
            f"<h1 style='margin: 0; font-size: 24px; color: #333;'><strong>{short_term_percentage}%</strong></h1>"
            f"<p style='font-size: 14px; margin: 0; color: #555;'>short-term rentals</p>"
            f"<h2 style='margin: 10px 0 0; font-size: 24px; color: #333;'><strong>{short_term_count} ({short_term_percentage}%)</strong></h2>"
            f"<p style='font-size: 14px; margin: 0; color: #555;'>short-term rentals</p>"
            f"<h2 style='margin: 10px 0 0; font-size: 24px; color: #333;'><strong>{longer_term_count} ({longer_term_percentage}%)</strong></h2>"
            f"<p style='font-size: 14px; margin: 0; color: #555;'>longer-term rentals</p>"
            "</div>"
          )
        
          display(HTML(output_html))
           
def plot_listings_per_host(selected_city, cities, listings_per_host_checkbox):
    with output_listings_per_host:
        output_listings_per_host.clear_output()  
        city_data = cities[selected_city]
        if listings_per_host_checkbox:
            city_data = city_data[city_data['calculated_host_listings_count'] > 1]
        city_data.loc[:, 'host_listings_count_capped'] = city_data['calculated_host_listings_count'].clip(upper=10)  # Cap the listings count at 10 for readability in the chart
        host_listings_counts = city_data['host_listings_count_capped'].value_counts().sort_index()
        # Create DataFrame for charting
        listings_per_host_data = pd.DataFrame({
            'listings_per_host': host_listings_counts.index,
            'count': host_listings_counts.values
        })
        listings_per_host_data['listings_per_host'] = listings_per_host_data['listings_per_host'].astype(str)
        listings_per_host_data.loc[listings_per_host_data['listings_per_host'] == '10', 'listings_per_host'] = '10+'
        x_order = [str(i) for i in range(1, 10)] + ['10+'] # Define order of x-axis labels
        listings_chart = alt.Chart(listings_per_host_data).mark_bar(color='#4A90E2', size=40).encode(
            x=alt.X('listings_per_host:O', title='Listings per Host', sort=x_order, axis=alt.Axis(labelAngle=0)),
            y=alt.Y('count:Q', title='Listings'),
            tooltip=[
                alt.Tooltip('listings_per_host:O', title='Listings per Host'),
                alt.Tooltip('count:Q', title='Count of Hosts')
            ]
        ).properties(
            width=500  
        )
        display(listings_chart)
    with output_description_listings:
        output_description_listings.clear_output()
        description = (
            "Some Airbnb hosts have multiple listings. <br> A host may list separate rooms in the same apartment, or multiple apartments<br> or homes available in their entirity."
            " Hosts with multiple listings are more likely to be running a business,<br> are unlikely to be living in the property, and in violation of most short term rental laws"
            "<br>designed to protect residential housing."
        )
        display(HTML(description))
    with output_listings_per_host_percentages:
        output_listings_per_host_percentages.clear_output()
        single_listing_count = city_data[city_data['calculated_host_listings_count'] == 1].shape[0]
        multi_listing_count = city_data[city_data['calculated_host_listings_count'] > 1].shape[0]
        total_listings = single_listing_count + multi_listing_count
        single_listing_percentage = (single_listing_count / total_listings * 100) if total_listings > 0 else 0
        multi_listing_percentage = (multi_listing_count / total_listings * 100) if total_listings > 0 else 0
        single_listing_percentage = round(single_listing_percentage, 1)
        multi_listing_percentage = round(multi_listing_percentage, 1)
        output_html = (
            f"<div style='text-align: center; font-family: Arial, sans-serif;'>"
            f"<h1 style='margin: 0; font-size: 24px; color: #333;'><strong>{multi_listing_percentage}%</strong></h1>"
            f"<p style='font-size: 14px; margin: 0; color: #555;'>multi-listings</p>"
            f"<h2 style='margin: 10px 0 0; font-size: 24px; color: #333;'><strong>{single_listing_count} ({single_listing_percentage}%)</strong></h2>"
            f"<p style='font-size: 14px; margin: 0; color: #555;'>single listings</p>"
            f"<h2 style='margin: 10px 0 0; font-size: 24px; color: #333;'><strong>{multi_listing_count} ({multi_listing_percentage}%)</strong></h2>"
            f"<p style='font-size: 14px; margin: 0; color: #555;'>multi-listings</p>"
            "</div>"
        )

        display(HTML(output_html))

def top_hosts_table(selected_city, cities):
    with output_hosts_table:
        output_hosts_table.clear_output()  
        city_data = cities[selected_city]
        city_summary = city_data.drop_duplicates(subset='host_name', keep='first')
        city_summary = city_summary[['host_name', 
                                         'calculated_host_listings_count_entire_homes', 
                                         'calculated_host_listings_count_private_rooms', 
                                         'calculated_host_listings_count_shared_rooms', 
                                         'calculated_host_listings_count']] 
        city_summary['#Hotel rooms'] = city_summary['calculated_host_listings_count'] - (
            city_summary['calculated_host_listings_count_entire_homes'] + 
            city_summary['calculated_host_listings_count_private_rooms'] + 
            city_summary['calculated_host_listings_count_shared_rooms']
        )
        
        city_summary.columns = [
            'Host Name', 
            '#Entire home/apts', 
            '#Private rooms', 
            '#Shared rooms', 
            '#Listings',
            '#Hotel rooms',
        ]
        city_summary = city_summary.sort_values(by='#Listings', ascending=False).reset_index(drop=True)
        city_summary_top49 = city_summary.head(50)
        display(city_summary_top49)
        
# Function to calculate the total listings based on selected filters and city data        
def calculate_total_listings(selected_city, only_entire_homes, activity_checkbox, license, short_term, listings_per_host):
  with output_total_listings:
        output_total_listings.clear_output()
        city_data = cities[selected_city]
        if only_entire_homes:
            city_data = city_data[city_data['room_type'] == 'Entire home/apt']
        if activity_checkbox:
            city_data.loc[:, 'last_review'] = pd.to_datetime(city_data['last_review'], errors='coerce')
            six_months_ago = datetime.now() - timedelta(days=180)
            city_data = city_data[(city_data['last_review'] >= six_months_ago) & (city_data['total_bookings'] > 60)]
        if license:
            city_data = city_data[city_data['license'].fillna(0) == 0] 
        if short_term:
             city_data = city_data[city_data['minimum_nights'] < 30]   
        if listings_per_host:
              city_data = city_data[city_data['calculated_host_listings_count'] > 1]
        total_listings_filtered = city_data.shape[0]
        total_listings_city = cities[selected_city].shape[0]
        percentage = (total_listings_filtered / total_listings_city * 100) if total_listings_city > 0 else 0
        percentage = round(percentage, 1)
        total_listings_html = (
            f"<div style='text-align: right;'>"
            f"<h2 style='margin: 0;'>{total_listings_filtered}</h2>"
            f"<p style='font-size: 14px; margin: 0;'>out of {total_listings_city} listings ({percentage}%)</p>"
            "</div>"
        )
        display(HTML(total_listings_html))
      
city_dropdown = Dropdown(
    options=list(cities.keys()),
    description='City:'
)

room_type_title = HTML("<h3 style='display:inline;'>Room Type</h3>")
only_entire_homes_checkbox = Checkbox(
    value=False,
    description='Only entire homes/apartments',
)

room_type_title_and_checkbox = HBox([
    room_type_title, 
    only_entire_homes_checkbox
], layout={'justify_content': 'space-between'})


activity_title = HTML("<h3 style='display:inline;'>Activity</h3>")
activity_checkbox = Checkbox(
    value=False,
    description="Only recent and frequently booked",
    layout={'width': 'auto'}  
)

activity_title_and_checkbox = HBox([
    activity_title,
    activity_checkbox
], layout={'justify_content': 'space-between'})

short_term_title =  HTML("<h3 style='display:inline;'>Short-Term Rentals</h3>")

short_term_checkbox = Checkbox(
    value=False,
    description="Only short-term rentals",
    layout={'width': 'auto'}  
)

short_term_title_and_checkbox = HBox([
    short_term_title,
    short_term_checkbox
], layout={'justify_content': 'space-between'})

line = HTML("<hr style='margin: 10px 0;'>")

short_term_layout = VBox([
    short_term_title_and_checkbox, line,
      HBox([
        output_description_short_term, 
        output_short_term_percentages
    ], layout={
        'justify_content': 'space-between', 
        'align_items': 'flex-start', 
        'width': '85%', 
        'margin': '10px'  
    }),  
    output_short_term
    ])

listings_per_host_title =  HTML("<h3 style='display:inline;'>Listings Per Host</h3>")
listings_per_host_checkbox = Checkbox(
    value=False,
    description="Only multi-listings",
    layout={'width': 'auto'}  
)

listings_per_host_title_and_checkbox = HBox([
    listings_per_host_title,
    listings_per_host_checkbox
], layout={'justify_content': 'space-between'})

listings_per_host_layout = VBox([
    listings_per_host_title_and_checkbox, line,
     HBox([
        output_description_listings, 
        output_listings_per_host_percentages
    ], layout={
        'justify_content': 'space-between', 
        'align_items': 'flex-start', 
        'width': '85%', 
        'margin': '10px'  
    }),  
    output_listings_per_host
    ])

license_title = HTML("<h3 style='display:inline;'>Licenses</h3>")
license_checkbox = Checkbox(
    value=False,
    description="Only unlicensed",
    layout={'width': 'auto'}  
)

license_title_and_checkbox = HBox([
    license_title,
    license_checkbox
], layout={'justify_content': 'space-between'})

top_hosts_title = HTML("<h3 style='display:inline;'>Top Hosts</h3>")

top_hosts_layout = VBox([
    top_hosts_title, line, output_hosts_table])

interactive_plot = interactive(plot_room_type, selected_city=city_dropdown, only_entire_homes=only_entire_homes_checkbox, activity=activity_checkbox, 
                               license=license_checkbox, short_term=short_term_checkbox, listings_per_host=listings_per_host_checkbox)

room_type_layout = VBox([
    room_type_title_and_checkbox, line, HBox([output_description, output_plot, output_percentages])
])

# Function to update all plots based on the current filters and selections
def update_plots(change):
    calculate_total_listings(city_dropdown.value, only_entire_homes_checkbox.value, activity_checkbox.value, license_checkbox.value, short_term_checkbox.value, listings_per_host_checkbox.value)
    plot_activity(city_dropdown.value, activity_checkbox.value)
    plot_room_type(city_dropdown.value, only_entire_homes_checkbox.value)
    plot_license(city_dropdown.value, cities, license_checkbox.value)
    plot_short_term(city_dropdown.value, cities, short_term_checkbox.value)
    plot_listings_per_host(city_dropdown.value, cities, listings_per_host_checkbox.value)
    top_hosts_table(city_dropdown.value, cities)

update_plots(None)

city_dropdown.observe(update_plots, names='value')
activity_checkbox.observe(update_plots, names='value')
only_entire_homes_checkbox.observe(update_plots, names='value')
license_checkbox.observe(update_plots, names='value')
short_term_checkbox.observe(update_plots, names='value')
listings_per_host_checkbox.observe(update_plots, names='value')

license_type_layout = VBox([
    license_title_and_checkbox, line,
     HBox([
        output_description_license, 
        output_percentages_license
    ], layout={
        'justify_content': 'space-between', 
        'align_items': 'flex-start', 
        'width': '85%', 
        'margin': '10px'  
    }),  
    output_license
])

activity_type_layout = VBox([
    activity_title_and_checkbox,  
    line,                         
    HBox([
        output_description_activity, 
        output_activity_averages
    ], layout={
        'justify_content': 'space-between', 
        'align_items': 'flex-start', 
        'width': '85%', 
        'margin': '10px'  
    }),  
    output_activity  
])

total_listings = HBox([
    output_total_listings
], layout={'justify_content': 'flex-start', 'align_items': 'center', 'margin': '-10px 0 8px auto'}) 

final_layout = VBox([
    city_dropdown,
    total_listings,
    room_type_layout,
    activity_type_layout,
    license_type_layout,
    short_term_layout,
    listings_per_host_layout,
    top_hosts_layout
]) 


display(final_layout)
```
